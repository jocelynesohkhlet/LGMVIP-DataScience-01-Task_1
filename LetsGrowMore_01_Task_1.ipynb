### Splitting Features and Target
<hr/>

```python
x= iris.drop(columns='Species', axis=1)
y= iris['Species']
```

```python
print(x)
```

```python
print(y)
```

### Splitting the data into Train & Test
<hr/>

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
X_train
```

```python
X_test
```

```python
print(x.shape,X_train.shape,X_test.shape)
```

#####  There are 150 rows in the origianal x data. 105 in the Train Dataset and 45 in the Test Dataset

### Model Building
<hr/>

#### Model 1: Support Vector Machine Algorithmn

```python
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(X_train,y_train)
```

```python
prediction1= model_svc.predict(X_test)

#Calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction1)*100)
```

##### Accuracy for this Model 1 is 95.56%.

#### Model 2 : Logistic Regression.

```python
from sklearn.linear_model import LogisticRegression

model_LR=LogisticRegression()
model_LR.fit(X_train,y_train)
```

```python
prediction2 = model_LR.predict(X_test)

#Calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction2)*100)
```

##### Accuracy for this Model 2 is 95.56%.

#### 3.Naive Bayes Algorithm

```python
from sklearn.naive_bayes import GaussianNB
accuracies={}
nb=GaussianNB()
nb.fit(X_train,y_train)
acc=nb.score(X_test,y_test)*100
accuracies['Naive Bayes']=acc
print("Accuracy of Naive Bayes:{:.2f}%".format(acc))
```

##### Accuracy for this Model 3 is 97.78%.

### Conclusion:
1. Model 1:Support Vector Machine Algorithm gives 95.56% Accuracy.
2. Model 2 :Logistic Regression gives 95.56% Accuracy.
3. Model 3 : Naive Bayes Algorithm gives 97.78%
